```python
lagos_fe = Lagos.copy()
```

```python
lagos_fe.columns
```

```python
## Price Per Neighborhood Relative to City
lagos_fe['city_avg_price'] = (
    lagos_fe.groupby('city')['price_ngn']
    .transform('median')
)
```

```python
lagos_fe['price_vs_city_avg'] = (
    lagos_fe['price_ngn'] /
    lagos_fe['city_avg_price']
)
```

```python
## Neighborhood Frequency
lagos_fe['neighborhood_frequency'] = (
    lagos_fe.groupby('neighborhood')['neighborhood']
    .transform('count')
)
```

```python
## City Frequency
lagos_fe['city_frequency'] = (
    lagos_fe.groupby('city')['city']
    .transform('count')
)
```

```python
## Premium Neighborhood
median_price = (
    lagos_fe.groupby('neighborhood')['price_ngn']
    .median()
)

threshold = median_price.quantile(.75)

premium = median_price[median_price > threshold].index

lagos_fe['premium_neighborhood'] = (
    lagos_fe['neighborhood']
    .isin(premium)
)
```

```python
## Luxury Property
luxury_threshold = lagos_fe['price_ngn'].quantile(.90)
lagos_fe['luxury_property'] = (
    lagos_fe['price_ngn'] >= luxury_threshold
)
```

```python
## Log Price
lagos_fe['log_price'] = np.log1p(lagos_fe['price_ngn'])
```

```python
## Number of Amenities
amenity_cols = [
    'serviced',
    'newly_built',
    'furnished',
    'has_bq'
]

for col in amenity_cols:
    lagos_fe[col] = (
        lagos_fe[col]
        .astype(str)
        .str.lower()
        .map({
            'yes':1,
            'true':1,
            '1':1,
            'no':0,
            'false':0,
            '0':0
        })
    )

lagos_fe['amenity_score'] = (
    lagos_fe[amenity_cols]
    .sum(axis=1)
)
```

```python
## Property Popularity
lagos_fe['property_type_frequency'] = (
    lagos_fe.groupby('property_type')['property_type']
    .transform('count')
)
```

```python
## Neighborhood Ranking
rank = (
    lagos_fe.groupby('neighborhood')['price_ngn']
    .median()
    .rank(ascending=False)
)

lagos_fe['neighborhood_rank'] = (
    lagos_fe['neighborhood']
    .map(rank)
)
```

```python
## Market Segment
def market_segment(price):

    if price < 500000:
        return "Affordable"

    elif price < 2000000:
        return "Mid-Market"

    elif price < 5000000:
        return "Premium"

    else:
        return "Luxury"

lagos_fe['market_segment'] = (
    lagos_fe['price_ngn']
    .apply(market_segment)
)
```

```python
## Importation of models
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
```

```python
from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error,
    r2_score
)
```

```python
## Creating Models
lr = LinearRegression()

dt = DecisionTreeRegressor(
    random_state=42
)

rf = RandomForestRegressor(
    n_estimators=200,
    random_state=42
)

gb = GradientBoostingRegressor(
    random_state=42
)
```

```python
## Training the Models
lr.fit(X_train_scaled, y_train)

dt.fit(X_train, y_train)

rf.fit(X_train, y_train)

gb.fit(X_train, y_train)
```

```python
## Make Predictions
lr_pred = lr.predict(X_test_scaled)

dt_pred = dt.predict(X_test)

rf_pred = rf.predict(X_test)

gb_pred = gb.predict(X_test)
```

```python
## Evaluation Function
def evaluate_model(name, y_true, y_pred):

    mae = mean_absolute_error(y_true, y_pred)

    rmse = np.sqrt(mean_squared_error(y_true, y_pred))

    r2 = r2_score(y_true, y_pred)

    return {
        "Model": name,
        "MAE": mae,
        "RMSE": rmse,
        "R² Score": r2
    }
```

```python
results = []

results.append(
    evaluate_model(
        "Linear Regression",
        y_test,
        lr_pred
    )
)

results.append(
    evaluate_model(
        "Decision Tree",
        y_test,
        dt_pred
    )
)

results.append(
    evaluate_model(
        "Random Forest",
        y_test,
        rf_pred
    )
)

results.append(
    evaluate_model(
        "Gradient Boosting",
        y_test,
        gb_pred
    )
)
```

```python
## Comparing Models
results_df = pd.DataFrame(results)

results_df.sort_values(
    "R² Score",
    ascending=False
)
```

```python
## visualizing the performance
plt.figure(figsize=(8,5))

sns.barplot(
    data=results_df,
    x='Model',
    y='R² Score',
    palette='viridis'
)

plt.title("Model Comparison")

plt.xticks(rotation=20)

plt.show()
```
